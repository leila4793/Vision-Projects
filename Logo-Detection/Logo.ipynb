```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image

# Reading the CSV file
file_path = "Top_20_Brands.csv"
df = pd.read_csv(file_path)

# Folder path where the images are located
image_folder = "Logos"  # Make sure this path is correct!

# Adding the full image path to the dataframe
df["filePath"] = df["fileName"].apply(lambda x: os.path.join(image_folder, x))

# List of top brands we found
top_brands = [
    "apple", "microsoft", "google", "amazon", "samsung", "tesla", "nike", "adidas",
    "toyota", "mercedes", "bmw", "disney", "pepsi", "visa", "mastercard", "netflix",
    "sony", "intel", "starbucks", "youtube"
]

# Filtering the dataframe for the selected brands
df_filtered = df[df['brand'].isin(top_brands)]

# Selecting a sample image for each brand
sample_images = df_filtered.groupby("brand").first().reset_index()

# Displaying the images
num_images = len(sample_images)
fig, axes = plt.subplots(4, 5, figsize=(15, 12))  # 4 rows and 5 columns
axes = axes.flatten()

for i, row in enumerate(sample_images.itertuples(index=False)):
    img_path = row.filePath  # Image file path
    if os.path.exists(img_path):  # Check if file exists
        try:
            img = Image.open(img_path)  # Open the image
            axes[i].imshow(img)
            axes[i].axis("off")  # Remove axes
            axes[i].set_title(row.brand.capitalize())  # Display brand name on the image
        except Exception as e:
            print(f"Error loading {img_path}: {e}")
    else:
        print(f"‚ùå Image for brand {row.brand} not found: {img_path}")

plt.tight_layout()
plt.show()
```

```python
import os
import pandas as pd
from shutil import copyfile

# 20 selected brands
top_20_brands = [
    "Google", "Adidas", "YouTube", "Nike", "Mercedes", "Samsung", "Apple", "Microsoft", 
    "Amazon", "Tesla", "Coca-Cola", "Pepsi", "BMW", "Toyota", "Disney", "McDonald's", 
    "Visa", "Mastercard", "Netflix", "Sony"
]

# Reading the main CSV file
file_path = "LogoDatabase_with_brands.csv"
df = pd.read_csv(file_path)

# Filtering the dataframe for the selected brands
selected_brands_df = df[df['brand'].str.contains('|'.join(top_20_brands), case=False, na=False)]

# Saving the selected brands in a new CSV file
selected_brands_df.to_csv('Top_20_Brands.csv', index=False)

# Folder path to store images
image_folder = "Logos"  # Folder where images are located
output_folder = "Selected_Brands_Images"  # Folder to store the selected images

# Creating the destination folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Moving images to the new folder
for _, row in selected_brands_df.iterrows():
    image_path = os.path.join(image_folder, row['fileName'])
    if os.path.exists(image_path):  # Checking if the file exists
        # Destination path
        destination_path = os.path.join(output_folder, row['fileName'])
        try:
            copyfile(image_path, destination_path)  # Copying the image to the new folder
            print(f"Image {row['fileName']} has been moved to the {output_folder} folder.")
        except Exception as e:
            print(f"Error while copying {row['fileName']}: {e}")
    else:
        print(f"Image {row['fileName']} not found.")
```

```python
import os
import requests
from bs4 import BeautifulSoup

# List of brands
top_20_brands = [
    "Google", "Adidas", "YouTube", "Nike", "Mercedes", "Samsung", "Apple", "Microsoft",
    "Amazon", "Tesla", "Coca-Cola", "Pepsi", "BMW", "Toyota", "Disney", "McDonald's",
    "Visa", "Mastercard", "Netflix", "Sony"
]

# Directory to save images
save_dir = "logos_of_brand"
os.makedirs(save_dir, exist_ok=True)

# Headers for the request (simulating a browser)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Function to get image URLs from Bing
def get_logo_urls(brand):
    search_url = f"https://www.bing.com/images/search?q={brand}+logo"
    response = requests.get(search_url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    
    img_urls = []
    # Collecting valid image URLs
    for img_tag in soup.find_all("img"):
        img_url = img_tag.get("src")
        if img_url and img_url.startswith(("http", "https")):
            img_urls.append(img_url)
        if len(img_urls) >= 70:  # For each brand, 70 images
            break
    return img_urls

# Function to download images from URLs
def download_images(urls, brand):
    if not urls:
        print(f"‚ùå No logos found for {brand}")
        return
    
    for i, url in enumerate(urls):
        response = requests.get(url, stream=True)
        if response.status_code == 200:
            file_path = os.path.join(save_dir, f"{brand}_{i+1}.png")
            with open(file_path, "wb") as file:
                for chunk in response.iter_content(1024):
                    file.write(chunk)
            print(f"‚úÖ {brand} logo {i+1} downloaded.")
        else:
            print(f"‚ùå Failed to download {brand} logo {i+1}.")

# Get and download 70 logos from each brand
for brand in top_20_brands:
    logo_urls = get_logo_urls(brand)
    download_images(logo_urls, brand)

```

```python
import os
import shutil
import pandas as pd

# Folder paths
logos_dir = "logos_of_brand"
selected_dir = "Selected_Brands_Images"
csv_file = "Top_20_Brands.csv"

# Load CSV file (if it exists)
if os.path.exists(csv_file):
    df = pd.read_csv(csv_file)
else:
    # If CSV file does not exist, create a new DataFrame
    df = pd.DataFrame(columns=['logoName', 'fileName', 'brand'])

# List of brands (for brand names in the images)
top_20_brands = [
    "Google", "Adidas", "YouTube", "Nike", "Mercedes", "Samsung", "Apple", "Microsoft",
    "Amazon", "Tesla", "Coca-Cola", "Pepsi", "BMW", "Toyota", "Disney", "McDonald's",
    "Visa", "Mastercard", "Netflix", "Sony"
]

# Adding images to the Selected_Brands_Images folder and CSV file
for brand in top_20_brands:
    # Images related to each brand
    brand_images = [f for f in os.listdir(logos_dir) if brand in f]
    
    for image in brand_images:
        # File paths
        source_path = os.path.join(logos_dir, image)
        destination_path = os.path.join(selected_dir, image)

        # Move image to the Selected_Brands_Images folder
        shutil.move(source_path, destination_path)
        print(f"‚úÖ {image} moved to {selected_dir}")
        
        # Create new data for CSV
        logo_name = os.path.splitext(image)[0]  # Logo name (without extension)
        file_name = image  # File name
        
        # Create a DataFrame for the new row
        new_data = pd.DataFrame({'logoName': [logo_name], 'fileName': [file_name], 'brand': [brand]})
        
        # Combine the new DataFrame with the main DataFrame
        df = pd.concat([df, new_data], ignore_index=True)

# Save updates to the CSV file
df.to_csv(csv_file, index=False)
print("‚úÖ CSV file updated successfully.")

```
