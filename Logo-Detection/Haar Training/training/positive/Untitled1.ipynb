```python
from PIL import Image
import os
```

```python
input_folder = "D:/my_class/AI/MvisionML/main/Logo-Detection/Haar Training/training/positive/rawdata"
output_folder = "D:/my_class/AI/MvisionML/main/Logo-Detection/Haar Training/training/positive/out"
```

```python
def convert_to_bmp(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    for filename in os.listdir(input_folder):
        input_path = os.path.join(input_folder, filename)
        
        try:
            with Image.open(input_path) as img:
                output_path = os.path.join(output_folder, os.path.splitext(filename)[0] + ".bmp")
                img.save(output_path, "BMP")
                print(f"Converted: {filename} -> {output_path}")
        except Exception as e:
            print(f"Error converting {filename}: {e}")

input_dir = input_folder
output_dir = output_folder  
convert_to_bmp(input_dir, output_dir)
```
